Este dataset passou por algumas modificações, o que tornou um pouco menos trabalhoso aplicá-lo em um modelo. Não tive muito o que fazer além de remover os valores ausentes (NA) que ainda podem ter ficado.

As modificações listadas no Kaggle foram:

1 - Foi feito um merge dos dados de treino e teste em um único arquivo CSV.
2 - Foram removidos 2 atributos, 'ticket' e 'cabin'.
3 - A coluna com as labels foi movida da primeira para a última coluna.
4 - Foram adicionadas colunas extras de zeros para entradas categóricas, a fim de torná-las mais adequadas para a codificação One-Hot.
5 - Os valores de 'sex' e 'embarked' foram substituídos por valores categóricos e binários, respectivamente.
6 - Valores faltantes em 'age' e 'fare' foram preenchidos com valores médios da época.

```python
import pandas as pd
from pandas.plotting import scatter_matrix
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv('dataset.csv')
dataset.describe()
```

```python
df = dataset.dropna('columns')
df.head(5)
```

```python
X=df.iloc[:,0:-1].values #dados
y=df.iloc[:,-1].values #dados
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)

```

```python
model = tree.DecisionTreeClassifier()
model.fit(X_train, y_train)
y_pred=model.predict(X_test)
acc = accuracy_score(y_test,y_pred)
acc
```
